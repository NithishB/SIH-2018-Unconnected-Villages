## Node, Edge and Graph classes

```python
class Node:
    def __init__(self,name,x,y,highway,pts,water,disaster,tourism):
        self.name = name
        self.x = x
        self.y = y
        self.highway = highway
        self.pts = pts
        self.water = water
        self.disaster = disaster
        self.tourism = tourism
        
    def displayNodeDetails(self):
        print("\nX: ",self.x,"  Y: ",self.y,"\nName : ",self.name,"\nHighway : ",self.highway,"\nPoints : ",self.pts,"\nWaterBodies : ",self.water,"\nDisaster : ",self.disaster, "\nTourism : ",self.tourism)
        
class Edge:
    def __init__(self,v1,v2,length, width,terrain, traffic, disaster):
        self.cost=0
        self.benefit=0
        self.weight=0
        self.v1 = v1
        self.v2 = v2
        self.length = length
        self.width = width
        self.terrain = terrain
        self.traffic = traffic
        self.disaster = disaster
        
    def displayEdgeDetails(self):
        print("\nv1 : ",self.v1.name,"\nv2 : ",self.v2.name,"\nweight : ",self.weight,"\nlength : ",self.length,"\nwidth : ",self.width,"\nterrain : ",self.terrain,"\ntraffic : ",self.traffic,"\ndisaster : ",self.disaster)

class Graph:
    def __init__(self):
        self.visited = []
        self.selected = []
        self.nodes = []
        self.edges = []
        self.vis = []
        
    def addNode(self,factors):
        temp = Node(factors[0],factors[1],factors[2],factors[3],factors[4],factors[5],factors[6],factors[7])
        self.nodes.append(temp)
    
    def addEdge(self,factors):
        for node in self.nodes:
            if node.name==factors[0]:
                v1=node
            if node.name==factors[1]:
                v2=node
        temp = Edge(v1,v2,factors[2],factors[3],factors[4],factors[5],factors[6])
        self.edges.append(temp)
    
    def listNodes(self):
        for node in self.nodes:
            node.displayNodeDetails()
    
    def listEdges(self,default=0):
        if(default==0):
            for edge in self.edges:
                edge.displayEdgeDetails()
        else:
            for edge in self.selected:
                edge.displayEdgeDetails()
    
    def optimize_cost(self):
        global factor
        for edge in self.edges:
            edge.cost = (factor[0]*edge.length) + (factor[1]*(edge.terrain)*(edge.length)) + (factor[2]*(edge.disaster)*(edge.length))
            edge.benefit =  (factor[3]*(edge.v1.highway+edge.v2.highway))+(factor[4]*(edge.v1.pts+edge.v2.pts)/2)+ (factor[5]*edge.traffic) + (factor[6]*(edge.v1.water+edge.v2.water))+ (factor[7]*(edge.v1.disaster+edge.v2.disaster)) + (factor[8] * edge.width) + (factor[9]*(edge.v1.tourism+edge.v2.tourism))    
            edge.weight = edge.cost/edge.benefit*100
            #print("\n\n",edge.v1.name,"-",edge.v2.name,":\nCost: ",edge.cost,"\nBenefit: ",edge.benefit,"\nWeight: ",edge.weight)
    
    def get_weight(self,x):
        return x.weight
    
    def sort_edges(self):
        self.edges = sorted(self.edges, key=self.get_weight)
    
    def findNeigh(self,i):
        neigh = []
        name = self.nodes[i].name
        for e in self.selected:
            if(e.v1.name == name):
                neigh.append(self.nodes.index(e.v2))
            elif(e.v2.name == name):
                neigh.append(self.nodes.index(e.v1))
        return neigh
    
    def isUtilCycle(self,cur_node,vis,parent):
        vis[cur_node] = True
        for i in self.findNeigh(cur_node):
            if  vis[i]==False : 
                if(self.isUtilCycle(i,vis,cur_node)):
                    return True
            elif  parent!=i:
                return True
        return False
    
    def isCyclic(self):
        vis = [False]*(len(self.nodes))
        for i in range(len(self.nodes)):
            if(vis[i]==False):
                if(self.isUtilCycle(i,vis,-1)):
                    return True
        return False
    
    def kruskal(self):
        self.sort_edges()
        for edge in self.edges:
            if(edge not in self.selected):
                edge.displayEdgeDetails()
                self.selected.append(edge)
                if(self.isCyclic()):
                    del self.selected[-1]
                
    def convertToGraph(self,obj):
        for i in range(len(obj.rPoints)):
            j=0
            while (j <= len(obj.rPoints[i][0])-1):
                self.addNode(['r'+str(i+1)+'p'+str(j+1),obj.rPoints[i][0][j][0],obj.rPoints[i][0][j][1],1,1,1,1,1])
                try:
                    self.addNode(['r'+str(i+1)+'p'+str(j+2),obj.rPoints[i][0][j+1][0],obj.rPoints[i][0][j+1][1],1,1,1,1,1])
                    self.addEdge(['r'+str(i+1)+'p'+str(j+1),'r'+str(i+1)+'p'+str(j+2),\
                                obj.dist(obj.rPoints[i][0][j][0],obj.rPoints[i][0][j][1],\
                                obj.rPoints[i][0][j+1][0],obj.rPoints[i][0][j+1][1]),0,0,0,0])
                    if(obj.rPoints[i][1] in [1,2]):
                        self.selected.append(self.edges[-1])
                except:
                    pass
                j+=1
        for i in range(len(obj.notConnected)):
            self.addNode(['v'+str(i+1),obj.notConnected[i][0][0],obj.notConnected[i][0][1],0,0,0,0,0])
```

## Shapefile Extraction and cleaning

```python
class shpOperations:
    
    def __init__(self,vpath,rpath):
        self.vil = shapefile.Reader(vpath)
        self.rds = shapefile.Reader(rpath)
        self.vPoints = []
        self.rPoints = []
        self.rRecords = []
        self.masterRoads = []
        self.masterPoints = []
        self.notConnected = []
        self.colorsRoad = ['Red', 'Green', 'Blue']
        self.colorsPoint = ['ko', 'k+', 'bo']
        self.allTypes = []
        self.extractPoints()
        
    def extractPoints(self):
        for i in range(len(self.vil.shapes())):
            self.vPoints.append(self.vil.shape(i).points[0])
        for i in range(len(self.rds.shapes())):
            self.rPoints.append(self.rds.shape(i).points)
        for i in range(len(self.rds.records())):
            lst = []
            for j in range(1,len(self.rds.fields)):
                try:
                    lst.append((self.rds.fields[j][0],self.rds.record(i)[j-1]))
                except:
                    pass
            self.rRecords.append(lst)
        for roadDetails in self.rRecords:
            if 'MR' in roadDetails[0][1]:
                self.allTypes.append(1)
            elif 'Minor' in roadDetails[0][1]:
                self.allTypes.append(2)
            elif 'Street' in roadDetails[0][1]:
                self.allTypes.append(3)
    
    def createMasterRepresentation(self):
        i = 0
        for road in self.rPoints:
            road = np.array(road)
            x = road[:,0]
            y = road[:,1]
            if self.allTypes[i] == 1:
                self.masterRoads.append((road, 1, 0))
            elif self.allTypes[i] == 2:
                self.masterRoads.append((road, 2, 0))
            elif self.allTypes[i] == 3:
                self.masterRoads.append((road, 3, 0))
            i = i+1
            
        for village in obj.vPoints:
            self.masterPoints.append((village, 0))
            
        self.masterRoads = np.array(self.masterRoads)
        self.masterPoints = np.array(self.masterPoints) 
        
    def checkforPoint(self,pt, A, B):
        flag = -1
        m = A[0]*(pt[1] - B[1]) + pt[0]*(B[1]- A[1]) + B[0]*(A[1] - pt[1])
        if m == 0:
            if(A[0] < B[0]):
                if(pt[0]>A[0] and pt[0] < B[0]):
                    flag = 1
                else:
                    return False
            else:
                if(pt[0]<=A[0] and pt[0] >= B[0]):
                    flag = 1
                else:
                    return False
            if(A[1] < B[1]):
                if(pt[1]>=A[1] and pt[1] <= B[1]):
                    return True
                else:
                    return False
            else:
                if(pt[1]<=A[1] and pt[1] >= B[1]):
                    return True
                else:
                    return False
        else:
            return False
    
    def findPointsConnected(self):
        for i in range(self.masterPoints.shape[0]):
            village = np.array(self.masterPoints[i][0])
            for road in self.masterRoads:
                for pointOnRoad in road[0]:
                    if (np.array_equal(pointOnRoad,village) and road[1]==1):
                        self.masterPoints[i][1] = 1
                    elif (np.array_equal(pointOnRoad,village) and road[1]==2):
                        self.masterPoints[i][1] = 1
                for j in range(len(road[0]) - 1):
                    if(self.checkforPoint(village,road[0][j], road[0][j+1])) and road[1] == 1:
                        self.masterPoints[i][1] = 1
                    elif(self.checkforPoint(village,road[0][j], road[0][j+1])) and road[1] == 2:
                        self.masterPoints[i][1] = 1
                        
    def plotRoadsandPoints(self,roads, points,roadLabels, pointLablels):
        plt.figure(figsize=(40,40))
        roadLabels = roadLabels.split(',')
        pointLabels = pointLablels.split(',')
        for road in roads:
            for i in range(len(roadLabels)):
                if(road[1] == int(roadLabels[i])):
                    plt.plot(road[0][:,0],road[0][:,1],color=self.colorsRoad[i],label=roadLabels[i])
        vn = 1
        for point in points:
            for i in range(len(pointLabels)):
                if(point[1] == int(pointLabels[i])):
                    plt.annotate(str(vn), xy=(point[0][0],point[0][1]), xytext=(point[0][0]+10,point[0][1]))
                    plt.plot(point[0][0],point[0][1],self.colorsPoint[i],label=pointLablels[i])
            vn+=1
        plt.show()
    
    def findUnconnected(self):
        notConnectedCount = 0
        for village in self.masterPoints:
            if village[1] == 0:
                self.notConnectedCount = notConnectedCount + 1
                self.notConnected.append((village))
        
        l = [18, 10, 29, 33, 36, 7, 30, 4, 13, 5, 6]
        l = sorted(l)
        for n in range(len(l)):
            self.notConnected = np.delete(self.notConnected,l[n],0)
            l = [h-1 for h in l]
    
    def distOfPointFromLine(self,pt, p1, p2):
        A = (p1[1] - p2[1])
        B = (p2[0] - p1[0])
        C = (p1[0]*p2[1] - p2[0]*p1[1])
        try:
            val = abs(A*pt[0] + B*pt[1] + C) / sqrt(A*A + B*B)
            x = (B*(B*pt[0]-A*p1[1])-A*C)/(A*A+B*B)
            y = (A*(A*pt[1]-B*pt[0])-B*C)/(A*A+B*B)
        except:
            pass
        return val, (x,y)
    
    def findIntersection(self,p1, p2, p3, p4):
        x = ((p1[0]*p2[1] - p1[1]*p2[0])*(p3[0]-p4[0]) - (p1[0]-p2[0])*(p3[0]*p4[1]-p3[1]*p4[0])) \
        / ((p1[0]-p2[0])*(p3[1]-p4[1])-(p1[1]-p2[1])*(p3[0]-p4[0]))

        y = ((p1[0]*p2[1] - p1[1]*p2[0])*(p3[1]-p4[1]) - (p1[1]-p2[1])*(p3[0]*p4[1]-p3[1]*p4[0])) \
        / ((p1[0]-p2[0])*(p3[1]-p4[1])-(p1[1]-p2[1])*(p3[0]-p4[0]))

        if(x >= min(p1[0],p2[0]) and x>= min(p3[0],p4[0]) and x<=max(p1[0],p2[0]) and x<= max(p3[0],p4[0]) and \
           y >= min(p1[1],p2[1]) and y>= min(p3[1],p4[1]) and y<=max(p1[1],p2[1]) and y<= max(p3[1],p4[1])):
            return (x,y)
        else:
            return None
        
    def dist(self,x1,y1,x2,y2):
        return np.sqrt((y2-y1)**2+(x2-x1)**2)
    
    def get_intersections(self):
        self.rPoints = self.masterRoads
        for i in range(len(self.rPoints)):
            if(self.rPoints[i][1] == 3):
                road1 = self.rPoints[i][0]
                for k in range(len(road1)-1):
                    pt1 = road1[k]
                    pt2 = road1[k+1]
                    for j in range(len(self.rPoints)):
                        if(i != j and self.rPoints[j][1] in [1,2,3]):
                            road2 = self.rPoints[j][0]
                            for l in range(len(road2)-1):
                                pt3 = road2[l]
                                pt4 = road2[l+1]
                                inter = self.findIntersection(pt1,pt2,pt3,pt4)
                                if(inter!=None):
                                    if(np.array(inter) not in self.rPoints[i][0]):
                                        self.rPoints[i][0] = np.insert(self.rPoints[i][0],np.where(road1 == pt1)[0][0]+1,np.array(inter), axis=0)
                                    if(np.array(inter) not in self.rPoints[j][0]):
                                        self.rPoints[j][0] = np.insert(self.rPoints[j][0],np.where(road2 == pt3)[0][0]+1,np.array(inter), axis=0)
    
    
    def calculateRoadLengths(self):
        for i in range(len(self.masterRoads)):
            road = self.masterRoads[i]
            l = 0
            for j in range(len(road[0])-1):
                l += np.sqrt((road[0][j+1][1]-road[0][j][1])**2 + (road[0][j+1][0]-road[0][j][0])**2)
            self.masterRoads[i][2] = l
```
